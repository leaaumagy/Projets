# **Analyse du cours de l'action d'une cryptomonnaie : l'Ethereum**

```python
import yahoofinancials as yf
from yahoofinancials import YahooFinancials
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

## **1. Récupération des données**

```python
yahoo_financials = YahooFinancials('ETH-USD')
data = yahoo_financials.get_historical_price_data(start_date='2020-01-01', 
                                                  end_date='2021-12-31', 
                                                  time_interval='monthly')
ETH_USD = pd.DataFrame(data['ETH-USD']['prices'])
ETH_USD = ETH_USD.drop('date', axis=1).set_index('formatted_date')
ETH_USD.head()
```

## **2. Génération d'un graphique du cours de l'action avec les courbes des taux d'ouverture, de fermeture, taux le plus haut, taux le plus bas**

```python
fig, ax = plt.subplots(figsize=(15,5))

#titre du graphique
plt.title("Stock prices for Ethereum (USD)")

#paramètres axe x
plt.gca().set_xlabel('Date')
plt.xticks(rotation='45')

#paramètres axe y
ax.set_ylabel('Price')
ax.set_ylim(100,5000)

fmt = '${x:.0f}'
tick = ticker.StrMethodFormatter(fmt)
ax.get_yaxis().set_major_formatter(tick)


#les différentes courbes a afficher
plt.plot(ETH_USD['open'],label='Open', color='steelblue')
plt.plot(ETH_USD['close'], label='Close', color='orange')
plt.plot(ETH_USD['low'], label='Low Price', linestyle=':', color='green')
plt.plot(ETH_USD['high'], label='High Price', linestyle=':', color='red')
plt.legend()


plt.show()
```
